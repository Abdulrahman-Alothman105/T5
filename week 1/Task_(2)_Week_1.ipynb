| Column Name | Description | Data Type |

# Tasks

```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
from sklearn.impute import SimpleImputer

df=pd.read_csv('/content/Task (2) Dataset.csv')
df.head()
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
df.duplicated().sum()
```

```python
imputer = SimpleImputer(strategy='mean')
imputer.fit(df[['TravelTimeHistoric', 'TravelTimeLive']])

df[['TravelTimeHistoric', 'TravelTimeLive']] = imputer.transform(df[['TravelTimeHistoric', 'TravelTimeLive']])
```

## 1. Exploratory Data Analysis (EDA)

### Perform summary statistics on the dataset.

```python
df.describe()
```

```python
df.describe(include='object')
```

### Identify and analyze patterns in the data.

### Visualize the distribution of key variables (e.g., Traffic_Index, Date).

```python
plt.figure(figsize=(8, 6))
sns.histplot(df['TrafficIndexLive'], kde=True)
plt.title('Distribution of Traffic Index Live')
plt.xlabel('Traffic Index Live')
plt.ylabel('Frequency')
plt.show()

plt.figure(figsize=(8, 6))
sns.histplot(df['TrafficIndexWeekAgo'], kde=True)
plt.title('Distribution of Traffic Index Week Ago')
plt.xlabel('Traffic Index Week Ago')
plt.ylabel('Frequency')
plt.show()

plt.figure(figsize=(8, 6))
sns.histplot(df['JamsCount'], kde=True)
plt.title('Distribution of Jams Count')
plt.xlabel('Jams Count')
plt.ylabel('Frequency')
plt.show()

plt.figure(figsize=(8, 6))
sns.histplot(df['JamsDelay'], kde=True)
plt.title('Distribution of Jams Delay')
plt.xlabel('Jams Delay')
plt.ylabel('Frequency')
plt.show()

plt.figure(figsize=(8, 6))
sns.histplot(df['JamsLength'], kde=True)
plt.title('Distribution of Jams Length')
plt.xlabel('Jams Length')
plt.ylabel('Frequency')
plt.show()

plt.figure(figsize=(8, 6))
sns.histplot(df['TravelTimeHistoric'], kde=True)
plt.title('Distribution of Travel Time Historic')
plt.xlabel('Travel Time Historic')
plt.ylabel('Frequency')
plt.show()

plt.figure(figsize=(8, 6))
sns.histplot(df['TravelTimeLive'], kde=True)
plt.title('Distribution of Travel Time Live')
plt.xlabel('Travel Time Live')
plt.ylabel('Frequency')
plt.show()
```

```python
df['Datetime'] = pd.to_datetime(df['Datetime'])
df['month']=df['Datetime'].dt.month

plt.figure(figsize=(8, 6))
sns.histplot(df['month'], kde=True)
plt.title('Distribution of Months')
plt.xlabel('Months')
plt.ylabel('Frequency')
plt.show()
```

## 2. Data Visualization

```python
sns.scatterplot(x='TrafficIndexLive', y='JamsCount', data=df)
plt.title('Traffic Index Live vs. Jams Count')
plt.show()

sns.scatterplot(x='TrafficIndexLive', y='JamsDelay', data=df)
plt.title('Traffic Index Live vs. Jams Delay')
plt.show()

sns.scatterplot(x='TrafficIndexLive', y='JamsLength', data=df)
plt.title('Traffic Index Live vs. Jams Length')
plt.show()

sns.scatterplot(x='TrafficIndexLive', y='TrafficIndexWeekAgo', data=df)
plt.title('Traffic Index Live vs. Traffic Index Week Ago')
plt.show()
```

* Ensure the visualizations are clear and informative.

### Create visualizations to illustrate the findings from the EDA.


```python
city_traffic = df.groupby('City')['TrafficIndexLive'].mean().sort_values(ascending=False)
plt.figure(figsize=(10, 6))
sns.barplot(x=city_traffic.index, y=city_traffic.values)
plt.title('Average Traffic Index by City')
plt.xlabel('City')
plt.ylabel('Average Traffic Index')
plt.xticks(rotation=90)
plt.show()


city_jams = df.groupby('City')['JamsCount'].mean().sort_values(ascending=False)
plt.figure(figsize=(10, 6))
sns.barplot(x=city_jams.index, y=city_jams.values)
plt.title('Average Jams Count by City')
plt.xlabel('City')
plt.ylabel('Average Jams Count')
plt.xticks(rotation=90)
plt.show()


city_jams_delay = df.groupby('City')['JamsDelay'].mean().sort_values(ascending=False)
plt.figure(figsize=(10, 6))
sns.barplot(x=city_jams_delay.index, y=city_jams_delay.values)
plt.title('Average Jams Delay by City')
plt.xlabel('City')
plt.ylabel('Average Jams Delay')
plt.xticks(rotation=90)
plt.show()

city_jams_length = df.groupby('City')['JamsLength'].mean().sort_values(ascending=False)
plt.figure(figsize=(10, 6))
sns.barplot(x=city_jams_length.index, y=city_jams_length.values)
plt.title('Average Jams Length by City')
plt.xlabel('City')
plt.ylabel('Average Jams Length')
plt.xticks(rotation=90)
plt.show()
```

### Use appropriate plots such as histograms, bar charts, pie charts, scatter plots, and heatmaps.

```python
df['Datetime'] = pd.to_datetime(df['Datetime'])
df['Date'] = df['Datetime'].dt.date
daily_traffic = df.groupby('Date')['TrafficIndexLive'].mean()
plt.figure(figsize=(12, 6))
plt.plot(daily_traffic.index, daily_traffic.values)
plt.title('Average Traffic Index Over Time')
plt.xlabel('Date')
plt.ylabel('Traffic Index')
plt.show()
```

```python
dfc=df[[ 'TrafficIndexLive'	,'JamsCount'	,'JamsDelay',	'JamsLength',	'TrafficIndexWeekAgo'	,'TravelTimeHistoric',	'TravelTimeLive']].corr()
print(dfc)
plt.figure(figsize=(10,8))
sns.heatmap(dfc,annot=True)
plt.title('correlation_matrix')
plt.show()
```
