```python
from sklearn.impute import SimpleImputer

numerical_cols = ['Urban_or_Rural_Area', 'Number_of_Casualties', 'Accident_Severity']

categorical_cols = ['Light_Conditions']

mean_imputer = SimpleImputer(strategy='mean')
df[numerical_cols] = mean_imputer.fit_transform(df[numerical_cols])

mode_imputer = SimpleImputer(strategy='most_frequent')
df[categorical_cols] = mode_imputer.fit_transform(df[categorical_cols])
```

```python
import numpy as np

def remove_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]

df=remove_outliers(df,'Location_Easting_OSGR')
df=remove_outliers(df,'Location_Northing_OSGR')
df=remove_outliers(df,'Number_of_Vehicles')
df=remove_outliers(df,'Number_of_Casualties')
df=remove_outliers(df,'1st_Road_Number')
df=remove_outliers(df,'2nd_Road_Number')
```

```python
from sklearn.preprocessing import OneHotEncoder

categorical_cols = ['Did_Police_Officer_Attend_Scene_of_Accident','Pedestrian_Crossing-Physical_Facilities','Road_Surface_Conditions','Light_Conditions']
one_hot_encoder = OneHotEncoder(sparse=False)
one_hot_encoded = one_hot_encoder.fit_transform(df[categorical_cols])
one_hot_encoded_df = pd.DataFrame(one_hot_encoded, columns=one_hot_encoder.get_feature_names_out(categorical_cols))

df=df.drop(columns=['Did_Police_Officer_Attend_Scene_of_Accident','Pedestrian_Crossing-Physical_Facilities','Road_Type','Weather_Conditions','Road_Surface_Conditions','Light_Conditions'])
df = pd.concat([df, one_hot_encoded_df], axis=1)
```

## Clustering
Apply K-Means clustering. Determine the optimal number of clusters and justify your choice.
* Find the `n_clusters` parameter using the elbow method.
* Train the model.

```python
from sklearn.cluster import KMeans

wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, random_state=42)
    kmeans.fit(df1)
    wcss.append(kmeans.inertia_)

plt.figure(figsize=(8, 5))
plt.plot(range(1, 11), wcss, marker='o')
plt.show()
```

## Evaluation
Evaluate the clustering result using appropriate metrics.


```python
kmeans = KMeans(n_clusters=3, random_state=42)

clusters = kmeans.fit_predict(df_pca)

df_pca_with_clusters = pd.DataFrame(df_pca, columns=[f'PC{i+1}' for i in range(pca.n_components_)])
df_pca_with_clusters['Cluster'] = clusters
```

```python
from sklearn.metrics import silhouette_score
silhouette_score(df_pca, kmeans.labels_)
```

## Plot the data points with their predicted cluster center

```python
plt.figure(figsize=(10, 7))

plt.scatter(df_pca_with_clusters['PC1'], df_pca_with_clusters['PC2'], c=df_pca_with_clusters['Cluster'], cmap='viridis', alpha=0.6)

centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, alpha=0.75, marker='X')

plt.show()
```
